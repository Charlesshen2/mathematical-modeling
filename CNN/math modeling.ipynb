```python
# 
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
ax = plt.axes(projection='3d')
z = np.linspace(0,100,1000)
x = np.sin(z)*z
y = np.cos(z)*z
ax.plot3D(x,y,z,'k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[0, 1, 2], [0, 1, 2]])
y = np.array([[0, 0, 0], [1, 1, 1]])


plt.plot(x, y,
         color='red',  # 
         marker='.',  # 
         linestyle='')  # 
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2])
y = np.array([0, 1])
x,y = np.meshgrid(x,y)


plt.plot(x, y,
         color='red',  # 
         marker='.',  # 
         linestyle='')  # 
plt.grid(True)
plt.show()
```
